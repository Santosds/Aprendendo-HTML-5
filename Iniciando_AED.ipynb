##Análise Exploratória de Dados(AED)

Fonte de dados: dados do e-commerce retirados do *Kaggle*.

```python
#importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Carregando o conjunto de dados
df = pd.read_csv('/content/Ecommerce_Consumer_Behavior_Analysis_Data.csv')
```

```python
#exibindo as primeiras linhas
print(df.head())
```

```python
#Informações gerais sobre o DataFrame
print(df.info())
```

```python
#Estatísticas descritivas
print(df.describe())
```

#Iniciando a limpeza e pré-processamento

Valores ausentes: NaN

Tipos de Dados;

Outliers;

```python
# Verificando valores ausentes
print(df.isnull().sum())
```

```python
# Verificando tipos de dados
print(df.dtypes)
```

```python
# Tratando valores ausentes (exemplo: imputação pela média)
df['Social_Media_Influence'].fillna(df['Social_Media_Influence'].mean(), inplace=True)
```

```python
# Corrigindo conflito de tipos de dados existentes na coluna Social Media Influence
df['Social_Media_Influence'] = pd.to_numeric(df['Social_Media_Influence'], errors='coerce')
#Realizando a tratativa após a correção dos erros de conflitos entre os tipos de dados na mesma coluna
df['Social_Media_Influence'].fillna(df['Social_Media_Influence'].mean(), inplace=True)
```

```python
# Convertendo tipos de dados(exemplo: string para datetime)
df['data'] = pd.to_datetime(df['data'])
```

#Análise Exploratória Visual

Distribuíção de Variáveis

Relações entre Variáveis

```python
# Histograma
sns.histplot(df['Age'])
plt.show()
```

```python
# Boxplot
sns.boxplot(x=df['Gender'], y=df['Age'])
plt.show()
```

```python
#Distribuíção de eventos por tipo
sns.countplot(x='Time_to_Decision', data=df)
plt.title('Distribuíção de Tipos de Eventos')
plt.show()
```

```python
# Correlação entre variáveis númericas
numerical_df = df.select_dtypes(include=['number'])
corr = numerical_df.corr()
sns.heatmap(corr, annot = True, cmap='coolwarm')
plt.title('Mapa de Calor de Correlação')
plt.show()
```
